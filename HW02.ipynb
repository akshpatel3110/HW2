**Q-1) Consider a data sample comprising of 10 data points as
given below.
The sample S ={1,0,2,3,6,1,2,8,1,3}**

a. Compute the mean value of the given sample S, and let that be
denoted by alpha

```python
import numpy as np                            # used numpy library to use math functions
import matplotlib.pyplot as plt               # used matplotlib to draw the graph

s = np.array([1,0,2,3,6,1,2,8,1,3])           # Given Sample dataset

slen = len(s)                                 # number of datapoints in dataset

alpha = np.mean(s)                            # mean value of sample s ie. alpha

print(f"The mean of the given sample S is {alpha}\n")  #printing the values
```

c. Now draw 100 bootstrap samples from S (with replacement).
Denote the properties of the i - th sample with the subscript i

```python
num_btsamples = 100                          # number of bootstrap samples is 100

boot_sample_means = []                       # create list to store btsamples values

raw_residules = []                           # list for raw residules

s_star = np.random.choice(s,size=slen)       # generating s star using random function
alpha_star = np.mean(s_star)

for i in range(num_btsamples):               # generating 100 Bootstrap samples using a for loop
    btsamples = np.random.choice(s,size=slen,replace=True)

    alpha_star_i = np.mean(btsamples)        # mean value of bootstrap samples
    boot_sample_means.append(alpha_star_i)

l_length = len(boot_sample_means)
print(l_length)                              #printing the values
print(boot_sample_means)
```
